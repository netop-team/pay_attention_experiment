```python
x = pd.read_pickle('data/winning_rate_llm/data.pkl')
```

```python
type(None) == dict
```

```python
win_dict = dict()

for k, v in x.items():
    win_count = 0
    total_count = 0
    total_no_null =0 
    total_none_type = 0

    for idx, row in v.T.iterrows():
        _, s= v.T.shape
        
        for i in range(s):
            row_dict = row[i]

            if row_dict is None:
                total_none_type += 1

            if type(row_dict) != dict: 
                continue
            
            print(row_dict)

            if row_dict['best_summary'] == 'modified':
                win_count += 1
                total_count += 1
                total_no_null += 1

            elif row_dict["best_summary"] == 'raw':
                total_count+= 1
                total_no_null += 1


    print(k, total_no_null, total_none_type)

    win_dict[k] = win_count/total_count
```
